*importing necessary libraries*

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

Function for plotting signals

```python
def plot_signal(t, x):
  '''
  t - Time vector for a Signal
  x - Signal 
  '''
  plt.figure(figsize=(18, 6))
  plt.plot(t, x)
  plt.show()
```

## Creating Message Signal

```python
'''
m  - amplitude sensitivity
M  - Amplitude of message signal
fm - Frequency of message signal
t  - Time vector for the signal
phase - Phase of message signal
mt - Modulating message signal
'''
A= 2
m = 1.0
M = A * m
t=np.arange(0, 50, 0.0001)
fm=.05
phase=0
mt=M*np.cos((2*np.pi*fm*t)+phase)
plot_signal(t, mt)
```

## Creating Carrier Signal

```python
'''
A  - Amplitude of carrier signal
fc - Frequency fo carrier signal
ct - Carrier signal
'''
phase = 0
fc=0.5 
ct=A*np.sin(2*np.pi*fc*t)
plot_signal(t, ct)
```

#**Amplitude modulation**

```python
'''
yat - Amplitude modulated signal
'''
yat=(1+m*np.cos((2*np.pi*fm*t) + phase))*A*np.sin(2*np.pi*fc*t);
plot_signal(t, yat)
```

#**Frequency Modulation**

```python
'''
yft - Frequency modulated signal
'''
yft = A*np.cos(2. * np.pi * (fc * t + m * np.sin(2*np.pi*fm*t)))
plot_signal(t, yft)
```

#**Phase Modulation**

###**Step 1: Creating Carrier Signal**

```python
'''
Ac - Amplitude of carrier signal
fc - Frequency fo carrier signal
t  - Time vector for the signal
phase - Phase of Carrier signal
ct - Carrier signal
'''
Ac=2; 
fc=2; 
t=np.arange(0, 3.0, 0.0001)
phase = 0
ct=Ac*np.sin(2*np.pi*fc*t + phase); 
plot_signal(t, ct)
```

###**Step 2: Creating message Signal**

```python
'''
k  - amplitude sensitivity
fm - Frequency of message signal
Am - Amplitude of the message signal
mt - Modulating message signal
'''
k = 1.0
fm=3
Am = 2
mt=Am*np.sin(2 * np.pi*fm*t)
plot_signal(t, mt)
```

###**Step 3: Computing Phase modulation**

```python
'''
yt - Phase modulated signal
'''
yt=Ac*np.sin((2*np.pi*fc*t) + (k * mt + phase))
plot_signal(t, yt)
```
